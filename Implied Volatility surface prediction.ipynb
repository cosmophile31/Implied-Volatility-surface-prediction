```python
import re
import matplotlib.pyplot as plt
import numpy as np

row = df.iloc[100]

strikes = []

for col in df.columns[2:]:
    numbers = re.findall(r'\d+', col)
    strike_raw = numbers[-1]
    strike = float(strike_raw[:-2])
    strikes.append(strike)

ivs = row[2:].values.astype(float)

valid = ~np.isnan(ivs)

strikes_clean = np.array(strikes)[valid]
ivs_clean = ivs[valid]

plt.figure(figsize=(14,5))

plt.plot(strikes_clean, ivs_clean, marker='o')

plt.title("Volatility Smile / Skew")
plt.xlabel("Strike")
plt.ylabel("Implied Volatility")

plt.grid(True)

plt.show()
```

```python
import re
import pandas as pd
import numpy as np

filled_df = df.copy()

iv_cols = df.columns[2:]

strike_map = {}

for col in iv_cols:
    numbers = re.findall(r'\d+', col)
    strike = float(numbers[-1][:-2])
    strike_map[col] = strike

numeric_surface = filled_df[iv_cols].copy()

numeric_surface.columns = [
    strike_map[col] for col in iv_cols
]

numeric_surface = numeric_surface.reindex(
    sorted(numeric_surface.columns),
    axis=1
)

numeric_surface = numeric_surface.interpolate(
    axis=1,
    method='cubic',
    limit_direction='both'
)

numeric_surface = numeric_surface.interpolate(
    axis=0,
    method='cubic',
    limit_direction='both'
)

numeric_surface = (
    numeric_surface
    .bfill()
    .ffill()
)

filled_df[iv_cols] = numeric_surface.values

print(filled_df.isnull().sum().sum())
```

# Validation Pipeline

```python
validation_df = df.copy()
```

```python
iv_cols = df.columns[2:]
```

```python
import numpy as np

np.random.seed(42)

mask = (
    np.random.rand(
        *validation_df[iv_cols].shape
    ) < 0.05
)
```

```python
true_values = validation_df[iv_cols].where(mask)
```

```python
validation_df[iv_cols] = (
    validation_df[iv_cols]
    .mask(mask)
)
```

```python
validation_df.isnull().sum().sum()
```

```python
reconstructed_df = validation_df.copy()
```

```python
import re

strike_map = {}

for col in iv_cols:
    
    numbers = re.findall(r'\d+', col)
    
    strike = float(numbers[-1][:-2])
    
    strike_map[col] = strike

numeric_surface = reconstructed_df[iv_cols].copy()

numeric_surface.columns = [
    strike_map[col] for col in iv_cols
]

numeric_surface = numeric_surface.reindex(
    sorted(numeric_surface.columns),
    axis=1
)

numeric_surface = numeric_surface.interpolate(
    axis=1,
    method='cubic',
    limit_direction='both'
)

numeric_surface = numeric_surface.interpolate(
    axis=0,
    method='linear',
    limit_direction='both'
)

numeric_surface = (
    numeric_surface
    .bfill()
    .ffill()
)

reconstructed_df[iv_cols] = numeric_surface.values
```

```python
predicted_values = (
    reconstructed_df[iv_cols]
    .where(mask)
)
```

```python
from sklearn.metrics import mean_squared_error

true_stack = true_values.stack()
pred_stack = predicted_values.stack()

common_index = true_stack.index.intersection(
    pred_stack.index
)

mse = mean_squared_error(
    true_stack.loc[common_index],
    pred_stack.loc[common_index]
)

print("Validation MSE:", mse)
```

```python
results = pd.DataFrame({
    'Method': ['Cubic Interpolation'],
    'Validation_MSE': [mse]
})

results
```

# Feature Engineering

```python
long_df = df.melt(
    id_vars=['datetime', 'underlying_price'],
    var_name='option',
    value_name='iv'
)
```

```python
long_df.head()
```

```python
import re

long_df['strike'] = (
    long_df['option']
    .str.extract(r'(\d+)')
    .astype(float)
)
```

```python
long_df['datetime'] = pd.to_datetime(
    long_df['datetime'],
    dayfirst=True
)

long_df['hour'] = (
    long_df['datetime'].dt.hour
)

long_df['minute'] = (
    long_df['datetime'].dt.minute
)
```

```python
long_df['moneyness'] = (
    long_df['strike'] /
    long_df['underlying_price']
)
```

```python
long_df.head()
```

```python
train_df = long_df[
    long_df['iv'].notnull()
].copy()
```
